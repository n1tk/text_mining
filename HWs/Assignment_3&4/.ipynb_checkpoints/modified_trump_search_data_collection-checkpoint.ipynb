```python
Trumpstring = ''.join(str(e) for e in text)
Trumpstring = re.sub(r"<a\S+", "", Trumpstring)
cleanTrumpstring = str(re.sub('\W+',',', Trumpstring))
keys = cleanTrumpstring.split(",")
```

```python
word_counts = {}
for word in keys:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1
```
